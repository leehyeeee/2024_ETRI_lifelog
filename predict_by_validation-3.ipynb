```python
# 컬럼 이름을 알파벳 순서대로 정렬
sorted_columns = sorted(x_train.columns)

# train 데이터프레임의 컬럼 순서를 정렬된 순서로 변경
x_train = x_train.reindex(columns=sorted_columns)

# test 데이터프레임의 컬럼 순서를 정렬된 순서로 변경
test = test.reindex(columns=sorted_columns)

```

## ※ 날짜 형식의 데이터는 학습시킬 때 오류난다

## 따라서 2월 1일 -> 32, 3월 1일 -> 61일 같이 바꿔주고 기존의 date은 drop시키겠다

```python
x_train['Month'] = x_train['date'].dt.month
x_train['Day_of_Month'] = x_train['date'].dt.day
test['Month'] = test['date'].dt.month
test['Day_of_Month'] = test['date'].dt.day

def to_days(x):
    month_to_days = {1:0, 2:31, 3:60, 4:91, 5:121, 6:152, 7:182, 8:213, 9:244, 10:274, 11:305, 12:335}
    return month_to_days[x]


x_train.loc[:, 'Day'] = x_train['Month'].apply(lambda x: to_days(x))
x_train['Day'] = x_train['Day'] + x_train['Day_of_Month']

test.loc[:, 'Day'] = test['Month'].apply(lambda x: to_days(x))
test['Day'] = test['Day'] + test['Day_of_Month']



print("Day Done.")
```

## 1-1. KNN

```python
answer_sample = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/데이터 경진대회/데이터/answer_sample.csv")
answer_sample = answer_sample.iloc[:, :2]
answer_sample
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
x_train_t = x_train.copy()
test_t = test.copy()
columns_to_drop = ['date', 'Month', 'Day_of_Month']

x_train_t.drop(columns=columns_to_drop, inplace=True)

columns_to_drop = ['date', 'Month', 'Day_of_Month']
test_t.drop(columns=columns_to_drop, inplace= True)

knn_model = KNeighborsClassifier(n_neighbors=20)
knn_model.fit(x_train_t, y_train)


pred = knn_model.predict(x_train_t)
print(metrics.f1_score(pred, y_train, average='macro'))


pred = knn_model.predict(test_t)

pred = pd.DataFrame(pred, columns = ['Q1', 'Q2', 'Q3', 'S1', 'S2', 'S3', 'S4'])
pred = pd.concat([answer_sample, pred], axis = 1)
pred.to_csv('/content/drive/MyDrive/Colab Notebooks/데이터 경진대회/데이터/answer_by_knn_only.csv', index=False)
```

## 1-2 knn + scaler + gridsearch

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

scaler1 = StandardScaler()
scaler1.fit(x_train_t) # 데이터 분리된 상황이면 train 데이터 기준으로
x_train_SS = scaler1.transform(x_train_t)
x_test_SS = scaler1.transform(test_t)



param_knn = {"n_neighbors":range(2,50,2)}
# GridSearchCV 함수 선언(교차검증도 같이 진행됨)
grid_search_knn = GridSearchCV(KNeighborsClassifier(),
                               param_knn,
                              cv = 10,
                              scoring='f1_micro'
                              )
grid_search_knn.fit(x_train_SS, y_train)

print("최적 파라미터 값: ", grid_search_knn.best_params_)
print("최고 점수: ", grid_search_knn.best_score_)

# 최고 성능 모델은 grid_search_knn 변수에 자동으로 저장됨
print("최고 성능 모델: ", grid_search_knn.best_estimator_)
knn_best = grid_search_knn.best_estimator_
pred = grid_search_knn.predict(x_test_SS)

pred = pd.DataFrame(pred, columns = ['Q1', 'Q2', 'Q3', 'S1', 'S2', 'S3', 'S4'])
pred = pd.concat([answer_sample, pred], axis = 1)
pred.to_csv('/content/drive/MyDrive/Colab Notebooks/데이터 경진대회/데이터/answer_by_grid_knn.csv', index=False)
```
