```python
import pandas as pd
import os
import numpy as np 
```

```python
def derived_variable(df):
    df1 = pd.DataFrame()
    
    if 'x' in df.columns: 
        
        df['mAcc'] = np.sqrt(df['x']*df['x'] + df['y']*df['y'] + df['z']*df['z'])
        df['date'] = pd.to_datetime(df['timestamp'], format='%Y-%m-%d').dt.date
        df1['mAcc_average'] = df.groupby(['date','subject_id'])['mAcc'].mean()
        df1['mAcc_min'] = df.groupby(['date','subject_id'])['mAcc'].min()
        df1['mAcc_max'] = df.groupby(['date','subject_id'])['mAcc'].max() 
        df1['mAcc_std'] = df.groupby(['date','subject_id'])['mAcc'].std()
             
    if 'heart_rate' in df.columns: 
        df['date'] = pd.to_datetime(df['timestamp'], format='%Y-%m-%d').dt.date
        df = df[df['heart_rate'] != 0]
        df1['wHr_average'] = df.groupby(['date','subject_id'])['heart_rate'].mean()
        df1['wHr_min'] = df.groupby(['date','subject_id'])['heart_rate'].min()
        df1['wHr_max'] = df.groupby(['date','subject_id'])['heart_rate'].max() 
        df1['wHr_std'] = df.groupby(['date','subject_id'])['heart_rate'].std() 
    
    df1.reset_index(inplace=True)
        
    return df1
```

# mAcc

```python
df_macc1 = pd.read_parquet('../data/test dataset/ch2024_test__m_acc_part_5.parquet')
df_macc1
```

```python
df_macc1 = derived_variable(df_macc1)
df_macc1.to_csv('../data/test_mAcc/df_macc5.csv')
```
