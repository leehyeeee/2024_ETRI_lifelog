# 공통이 아닌 컬럼 예측 복원

```python
import pandas as pd
```

### train

```python
train = pd.read_csv('../data/train_total1.csv', index_col=0) 
train_gps = pd.read_csv('../data/train_gps_distance.csv', index_col=0)
# train_val = pd.read_csv('../data/train_val_set.csv')

display(train.head(3), train_gps.head(3))
```

```python
train['WALKING'] = 0
train['RUNNING'] = 0
train1 = train[['date','month','weekday','weekend', 'userId', 
                'activity', 'IN_VEHICLE',	'ON_BICYCLE',	'STILL',	'UNKNOWN', 'WALKING', 'RUNNING',
                'caffeine_y_caffeinated drink','alcohol_y_0','alcohol_y_wine','caffeine_y_coffee','caffeine_y_0',
                'e4Eda_min','e4Temp_max','pmEmotion_y','pmStress_y', 
                'Hr_average','Hr_max', 'Hr_min','Hr_std',
                'mAcc_average','mAcc_max', 'mAcc_min','mAcc_std']] # gps 불포함 데이터
train1 = train1.rename(columns={'userId':'subject_id'})
train2 = pd.merge(train1, train_gps, on=['subject_id', 'date']) # gps 포함 데이터

display(train1.head(),train2.head())
```

### val

```python
val = pd.read_csv('../data/음성_컬럼수_줄인_val.csv', index_col=0)
val_mAcc = pd.read_csv('../data/val_mAcc.csv', index_col=0)
val_wHr = pd.read_csv('../data/val_wHr.csv', index_col=0) 
val_gps = pd.read_csv('../data/val_gps_distance.csv', index_col=0)

display(val.head(3),val_mAcc.head(3),val_wHr.head(3), val_gps.head(3))
```

```python
val1 = val[['date','month', 'weekday','weekend','subject_id', 
            'm_activity', 'IN_VEHICLE',	'ON_BICYCLE',	'STILL',	'UNKNOWN',	'RUNNING', 'WALKING',
            'Nature Sounds', 'Environmental Sounds', 'Human Activity Sounds', 'Machine and Traffic Sounds', 'Alert and Alarm Sounds', 'Music and Instrument Sounds', 'Other']]

val1 = pd.merge(val1, val_mAcc, on=['date','subject_id'])
val1 = pd.merge(val1, val_wHr, on=['date','subject_id']) # gps 불포함 데이터

val2 = pd.merge(val1, val_gps, on=['date','subject_id']) # gps 포함 데이터

display(val1.head(), val2.head())
```

### test

```python
test = pd.read_csv('../data/음성_컬럼수_줄인_test.csv', index_col=0)
test_mAcc = pd.read_csv('../data/test_mAcc.csv', index_col=0)
test_wHr = pd.read_csv('../data/test_wHr.csv', index_col=0) 
test_gps = pd.read_csv('../data/test_gps_distance.csv', index_col=0)

display(test.head(3), test_mAcc.head(3),test_wHr.head(3),test_gps.head(3))
```

```python
test1 = test[['date','month', 'weekday','weekend','subject_id', 
            'm_activity','IN_VEHICLE',	'ON_BICYCLE',	'STILL',	'UNKNOWN',	'RUNNING', 'WALKING',
            'Nature Sounds', 'Environmental Sounds', 'Human Activity Sounds', 'Machine and Traffic Sounds', 'Alert and Alarm Sounds', 'Music and Instrument Sounds', 'Other']]

test1 = pd.merge(test1, test_mAcc, on=['date','subject_id'])
test1 = pd.merge(test1, test_wHr, on=['date','subject_id']) # gps 불포함 데이터
test2 = pd.merge(test1, test_gps, on=['date','subject_id'], how='outer') # gps 포함 데이터

display(test1.head(), test2.head())
```

**train, val, test 공통 데이터**

- 'date','userId'
- 'activity', 'IN_VEHICLE', 'ON_BICYCLE', 'RUNNING', 'STILL', 'UNKNOWN', 'WALKING', 'ON_FOOT'
- 'e4Hr_average','e4Hr_max', 'e4Hr_min','e4Hr_stv'
- 'mAcc_average','mAcc_max', 'mAcc_min','mAcc_std'
- mGps -> 이동거리 gps_distance

**공통이 아닌 데이터**

- train : 설문, 수면 데이터

'caffeine_y_caffeinated drink','alcohol_y_0','alcohol_y_wine','caffeine_y_coffee','caffeine_y_0', 'e4Eda_min','e4Temp_max','pmEmotion_y','pmStress_y'

- val&test : 소리 데이터

'Nature Sounds', 'Environmental Sounds', 'Human Activity Sounds', 'Machine and Traffic Sounds', 'Alert and Alarm Sounds', 'Music and Instrument Sounds', 'Other'

```python
def val_test_col_modify(df):
    df = df.rename(columns={'m_activity':'activity'})

    df['str'] = 'User'
    df['subject_id'] = df['subject_id'].astype(str) 
    df['subject_id'] = df['str'] + df['subject_id']
    df = df.drop(labels='str', axis=1)
    
    return df  
```

```python
val1 = val_test_col_modify(val1)
test1 = val_test_col_modify(test1)
display(val1.head(),test1.head())
```

```python
val2 = val_test_col_modify(val2)
test2 = val_test_col_modify(test2)
display(val2.head(),test2.head())
```

## 피쳐 컬럼 예측을 위한 데이터 통합

### train -> val

```python
# gps 불포함
intersection_col1 = list(set(train1.columns) & set(val1.columns))
to_predict_val1 = pd.concat([train1, val1[intersection_col1]], axis = 0) 

# gps 포함
intersection_col2 = list(set(train2.columns) & set(val2.columns))
to_predict_val2 = pd.concat([train2, val2[intersection_col2]], axis = 0) 
```

### train -> test

```python
# gps 불포함
intersection_col1 = list(set(train1.columns) & set(test1.columns))
to_predict_test1 = pd.concat([train1, test1[intersection_col1]], axis = 0) 

# gps 포함
intersection_col2 = list(set(train2.columns) & set(test2.columns))
to_predict_test2 = pd.concat([train2, test2[intersection_col2]], axis = 0) 
```

### val&test -> train

```python
# gps 불포함
intersection_col1 = list(set(val1.columns) & set(train1.columns))
val_test1 = pd.concat([val1, test1], axis = 0) 
to_predict_train1 = pd.concat([val_test1, train1[intersection_col1]], axis = 0) 

# gps 포함
intersection_col2 = list(set(val2.columns) & set(train2.columns))
val_test2 = pd.concat([val2, test2], axis = 0) 
to_predict_train2 = pd.concat([val_test2, train2[intersection_col2]], axis = 0) 
```

### 예측을 위한 데이터 저장

```python
to_predict_val1.to_csv('../data/df_to_predict_val1.csv')
to_predict_val2.to_csv('../data/df_to_predict_val2.csv')

to_predict_test1.to_csv('../data/df_to_predict_test1.csv')
to_predict_test2.to_csv('../data/df_to_predict_test2.csv')

to_predict_train1.to_csv('../data/df_to_predict_train1.csv')
to_predict_train2.to_csv('../data/df_to_predict_train2.csv')
```
