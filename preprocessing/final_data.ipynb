```python
import pandas as pd
import numpy as np

val1 = pd.read_csv("../data/음성_컬럼수_줄인_val.csv", index_col=0)
test1 = pd.read_csv("../data/음성_컬럼수_줄인_test.csv", index_col=0)
val2 = pd.read_csv('../data/val_final_v2.csv', index_col=0)
test2 = pd.read_csv('../data/test_final_v2.csv', index_col=0)

display(val1.head(),val2.head(),test1.head(),test2.head())
```

```python
val1['subject_id'] = val1['subject_id'].astype(str)
val1['subject_id'] = 'User' + val1['subject_id']

test1['subject_id'] = test1['subject_id'].astype(str)
test1['subject_id'] = 'User' + test1['subject_id']
```

```python
val = pd.merge(val1, val2, on=list(set(val1.columns)&set(val2.columns)))
test = pd.merge(test1, test2, on=list(set(test1.columns)&set(test2.columns)))

display(val.head(),test.head())
```

```python
val.to_csv('../data/final_val.csv')
test.to_csv('../data/final_test.csv')
```
