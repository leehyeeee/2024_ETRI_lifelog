```python
import os
import pandas as pd
import numpy as np
from datetime import datetime
```

## timestamp_label

- ts			timestamp
- action (16)		sleep, personal_care, work, study, household, care_housemem (caregiving), recreation_media, entertainment, 
			outdoor_act (sports), hobby, recreation_etc (free time), shop, communitiy_interaction (regular activity), travel, meal, socialising
- actionOption (73)		Details of the selected action. See the description below.
- actionSub		meal_amount when action=meal or snack
			move_method when action=travel
- actionSubOption		1 (light), 2 (moderate), 3 (heavy) when actionSub=meal_amount
			1 (walk), 2 (driving), 3 (taxi, passenger), 4 (personal mobility), 5 (bus), 6 (train, subway), 7 (others) when actionSub=move_method
- condition		ALONE, WITH_ONE, WITH_MANY
- conditionSub1Option	1 (with families), 2 (with friends), 3 (with colleagues), 4 (acquaintances), 5 (others)
- conditionSub2Option	1 (passive in conversation), 2 (moderate participation in conversation), 3 (active in conversation)
- place			home, workplace, restaurant, outdoor, other_indoor
- emotionPositive (Valence)	(negative) 1-2-3-4-5-6-7 (positive)
- emotionTension (Arousal)	(relaxed) 1-2-3-4-5-6-7 (aroused)
- activity***		0 (IN_VEHICLE), 1 (ON_BICYCLE), 2 (ON_FOOT), 3 (STILL), 4 (UNKNOWN), 5 (TILTING), 7 (WALKING), 8 (RUNNING)

```python
timestamp_label = pd.DataFrame()

directory1 = '../data/users/'
file_list1 = os.listdir(directory1)
    
for i in range(0, len(file_list1)):
    directory2 = directory1 + '/' + file_list1[i] #../data/users/user01/
    file_list2 = os.listdir(directory2)
        
    num = directory2.split('/')[4]
        
    for j in range(0,len(file_list2)):
        directory3 = directory2 + '/' + file_list2[j] #/data/users/user01/00000000
        file_list3 = os.listdir(directory3)
        label_directory = directory3 + '/' + file_list3[0] 
            
        df = pd.read_csv(label_directory)
        df['userId'] = num
        df['date'] = datetime.fromtimestamp(int(file_list2[j])).strftime('%Y-%m-%d')
        timestamp_label = pd.concat([timestamp_label, df], ignore_index=True)
        
timestamp_label.to_csv('../data/timestamp_label.csv',index=False)
```

```python
list = ['sleep', 'personal_care', 'work', 'study', 'household', 'care_housemem', 'recreation_media', 'entertainment', 
			'outdoor_act', 'hobby', 'recreation_etc', 'shop', 'community_interaction', 'travel', 'meal', 'socialising']

n = 0
for i in list:
    action = timestamp_label[timestamp_label['action']==i]['actionOption']
    print(i,':', action.unique(), action.nunique())
    n += action.nunique()
    
print(n)
```

1) **action**

action에서 recreation_media와 entertainment, hobby 유사해보임

action에서 household와 care_housemem 유사해보임

2) **actionOption**

actionOption 중 데이터가 없는 것도 존재함 -> 65/73

[actionOption 중 유의미해보이는 것]

<action = 'sleep'> 111 sleep, 112 sleepless 

<action = 'recreation_etc'> 792 Smoking, 793 Do nothing and rest

```python
# action 컬럼 정리(sleep 컬럼은 actionOption 사용하기 때문에 제거)
df_action = timestamp_label.groupby(['userId', 'date'])['action'].value_counts().unstack(fill_value=0).reset_index()
df_action = df_action.add_prefix('action_')
df_action.rename(columns={'action_userId':'userId','action_date':'date'}, inplace=True)
df_action = df_action.drop(labels='action_sleep', axis=1)
df_action
```

```python
# actionOption 정리
df_actionOp = timestamp_label.groupby(['userId', 'date'])['actionOption'].value_counts().unstack(fill_value=0).reset_index()[['userId','date',111,112, 792, 793]] # 111:sleep, 112:sleepless
df_actionOp.rename(columns={111:'actOp_sleep',112:'actOp_sleepless', 792:'actOp_smoking', 793:'actOp_rest'}, inplace=True)
df_actionOp
```

### condition
- condition		

ALONE, WITH_ONE, WITH_MANY
- conditionSub1Option	

1 (with families), 2 (with friends), 3 (with colleagues), 4 (acquaintances), 5 (others)
- conditionSub2Option	

1 (passive in conversation), 2 (moderate participation in conversation), 3 (active in conversation)

```python
df = timestamp_label[['userId','date','condition','conditionSub1Option','conditionSub2Option']].copy()
df = df.astype({'conditionSub1Option':'string', 'conditionSub2Option':'string'})
df= df.fillna('0')
df.info()
```

```python
cond1_dic = {'1.0':'with_families', '2.0':'with_friends', '3.0':'with_colleagues','4.0':'with_acquatintances','5.0':'with_others', '0':'cond1_NaN'}
cond2_dic = {'1.0':'passive_conversation','2.0':'moderate_conversation','3.0':'active_conversation','0':'cond2_NAN'}

df['conditionSub1Option'] =df['conditionSub1Option'].map(cond1_dic)
df['conditionSub2Option'] =df['conditionSub2Option'].map(cond2_dic)
df
```

```python
cond_count = df.groupby(['userId', 'date'])['condition'].value_counts().unstack(fill_value=0).reset_index()
cond1_count = df.groupby(['userId', 'date'])['conditionSub1Option'].value_counts().unstack(fill_value=0).reset_index()
cond2_count = df.groupby(['userId', 'date'])['conditionSub2Option'].value_counts().unstack(fill_value=0).reset_index()

cond_fin = pd.merge(cond_count, cond1_count, on=['userId','date'])
cond_fin = pd.merge(cond_fin, cond2_count, on=['userId','date'])
cond_fin
```

### activity 

- activity***		

0 (IN_VEHICLE), 1 (ON_BICYCLE), 2 (ON_FOOT), 3 (STILL), 4 (UNKNOWN), 5 (TILTING), 7 (WALKING), 8 (RUNNING)

```python
df_new = timestamp_label[['userId','date','activity']].groupby(['userId', 'date'])['activity'].count().reset_index()
df_new.head()
```

```python
# subject_id, Date로 묶어준다음 0~8번 활동을 하루에 각각 몇 번씩 했는지 알기 위해 groupby 사용
activity_count = timestamp_label.groupby(['userId', 'date'])['activity'].value_counts().unstack(fill_value=0)
activity_count
```

```python
# 0 (IN_VEHICLE), 1 (ON_BICYCLE), 2 (ON_FOOT), 3 (STILL), 4 (UNKNOWN), 5 (TILTING), 7 (WALKING), 8 (RUNNING)
```

```python
# 컬럼을 0~8 대신에 숫자에 해당하는 활동명으로 바꿔줌
activity_count.columns = ['IN_VEHICLE', 'ON_BICYCLE', 'ON_FOOT', 'STILL','UNKNOWN', 'other']
```

```python
# 총 활동 횟수 df + 하루에 각 활동을 몇 번 했는지 알려주는 df
df_mact_fin = df_new.merge(activity_count, on=['userId', 'date'], how='left')
df_mact_fin
```
